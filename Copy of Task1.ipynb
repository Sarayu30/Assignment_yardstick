```python


from yt_dlp import YoutubeDL
import os # Import the os module


def download_youtube_video_yt_dlp(video_url, download_path):
    ydl_opts = {
        'format': 'best',  # Download the best quality video
        'outtmpl': os.path.join(download_path, '%(title)s.%(ext)s'),  # Save with the title as filename
        'nocheckcertificate': True  # Disable SSL certificate validation
    }

    with YoutubeDL(ydl_opts) as ydl:
        try:
            ydl.download([video_url])
            video_info = ydl.extract_info(video_url, download=False)
            video_filename = os.path.join(download_path, f"{video_info['title']}.mp4")
            return video_filename
        except Exception as e:
            print(f"An error occurred: {e}")
            return None


video_url = 'https://www.youtube.com/watch?v=D00BapLQRuI'  #  video URL
download_path = '/content/drive/My Drive/Test'

video_file_path = download_youtube_video_yt_dlp(video_url, download_path)
if video_file_path:
    print(f"Video downloaded and saved as: {video_file_path}")
else:
    print("Failed to download the video.")
```

# Extract the audio from the video

```python
from yt_dlp import YoutubeDL
import os
from moviepy.editor import VideoFileClip



def extract_audio(video_file_path):
    if video_file_path:
        # Specify the output audio file path
        audio_file_path = os.path.splitext(video_file_path)[0] + '.mp3'

        # Load the video file and extract audio
        video = VideoFileClip(video_file_path)
        video.audio.write_audiofile(audio_file_path)

        # Close the video clip
        video.close()

        return audio_file_path
    else:
        print("No video file to extract audio from.")
        return None

# Example usage
video_url = 'https://www.youtube.com/watch?v=D00BapLQRuI'  # Replace with your video URL
download_path = '/content/drive/My Drive/Test'  # Update to your Google Drive path

# Download the video
video_file_path = download_youtube_video_yt_dlp(video_url, download_path)

# Extract audio from the downloaded video
audio_file_path = extract_audio(video_file_path)
if audio_file_path:
    print(f"Audio extracted and saved as: {audio_file_path}")
else:
    print("Failed to extract audio.")
```
